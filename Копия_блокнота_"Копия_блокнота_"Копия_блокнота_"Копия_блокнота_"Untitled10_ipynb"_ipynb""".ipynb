```python
def same_letter_set(word1, word2, word3):
  set1 = set(word1.lower())
  set2 = set(word2.lower())
  set3 = set(word3.lower())
  return set1 == set2 == set3
print("Введите три слова:")
word1 = input("Первое слово: ")
word2 = input("Второе слово: ")
word3 = input("Третье слово: ")
if same_letter_set(word1, word2, word3):
  print("Да, для записи всех трёх слов использован одинаковый набор букв.")
else:
 print("Нет, наборы букв в словах различаются.")
```

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Загружаем встроенный датасет
california = fetch_california_housing()

# Создаём DataFrame из признаков
df_california = pd.DataFrame(california.data, columns=california.feature_names)

# Добавляем целевую переменную (медианная стоимость дома) в DataFrame
df_california['MedHouseVal'] = california.target

# Находим минимальное и максимальное значения медианной стоимости
min_value = df_california['MedHouseVal'].min()
max_value = df_california['MedHouseVal'].max()

print(f"Минимальное значение медианной стоимости дома: {min_value}")
print(f"Максимальное значение медианной стоимости дома: {max_value}")
```
